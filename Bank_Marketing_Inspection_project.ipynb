# Bank Marketing Data Analysis
This notebook presents an analysis of the `bankmarketing.csv` dataset, which includes data related to a bank's marketing campaigns. The main goal is to understand customer behavior and predict whether a client will subscribe to a term deposit.

```python

import pandas as pd

# Load the dataset
df = pd.read_csv('bankmarketing.csv')

# Display the first few rows
df.head()
```

```python
import pandas as pd
# Load the dataset
df = pd.read_csv('bankmarketing.csv')

# Display the first few rows
df.head()
```

```python

# Check for missing values and data types
df.isnull().count()
```

```python

# Summary statistics
df.describe()
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Plot the distribution of the target variable
df['y'].value_counts().plot(kind='bar')

plt.title('Distribution of Target Variable (Term Deposit Subscription)')
plt.xlabel('Subscription')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python

# Correlation matrix for numerical features
numerical_features = df.select_dtypes(include=['int64', 'float64'])
corr_matrix = numerical_features.corr()
print (corr_matrix)
```

```python
sns.heatmap(corr_matrix,
            annot=True,
            cmap='coolwarm',
            fmt='.2f')
plt.title("Correlation Matrix of Numerical Features")
plt.show
```
